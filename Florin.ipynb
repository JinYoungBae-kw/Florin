#### 디렉토리 내 파일 갯수 확인

```python
# 디렉토리 경로 설정

def count_files(folder_path):
    files = os.listdir(folder_path)
    file_count = len([f for f in files if os.path.isfile(os.path.join(folder_path, f))])
    print(f"디렉토리 {folder_path}에 있는 파일 수: {file_count}개")

count_files("/content/output/inference_result/labels")
```

#### 폴더 내 파일 삭제

```python
import os

def delete_all_files(folder_path):
    if not os.path.exists(folder_path):
        print(f"폴더가 존재하지 않습니다: {folder_path}")
        return

    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):  # 파일인지 확인
            os.remove(file_path)  # 파일 삭제
            print(f"삭제 완료: {file_path}")

delete_all_files("/content/output/check_result")
```

#### yolo_dataset 폴더에 원본 이미지 복사

```python
import shutil
import os
from glob import glob

# 데이터셋 경로 설정
base_dataset_dir = "/content/gdrive/MyDrive/base_dataset"
base_yolo_dir = "/content/yolo_dataset"

for dir_name in ["train", "val", "test"]:
    # 원본 이미지 경로
    base_images_dir = os.path.join(base_dataset_dir, dir_name, "images")
    # yolo 이미지 경로
    yolo_images_dir = os.path.join(base_yolo_dir, dir_name, "images")
    # 이미지 파일 이름 리스트
    image_files = []
    for extension in ["*.jpg", "*.jpeg", "*.JPG", "*.JPEG"]:
        image_files.extend(glob(os.path.join(base_images_dir, extension)))


    for image_file in image_files:
        # 복사하는 이미지 최종 경로
        yolo_images = os.path.join(yolo_images_dir, os.path.basename(image_file))

        # 이미지 복사
        shutil.copy(image_file, yolo_images)
        print(f"복사 완료: {yolo_images}")
```

#### yolo_label 파일 생성

```python
import json
import re

# 클래스 매핑
CLASS_MAPPING = {0: 0, 1: 1, 2: 2, 13: 3, 14: 4, 16: 5, 17: 6, 18: 7}

# 데이터셋 유형별 변환 수행 (train, val, test)
for dir_name in ["train", "val", "test"]:
    labels_dir = os.path.join(base_dataset_dir, dir_name, "labels")
    images_dir = os.path.join(base_dataset_dir, dir_name, "images")
    output_dir = os.path.join(base_yolo_dir, dir_name, "labels")

    # JSON 파일 리스트 가져오기
    json_files = glob(os.path.join(labels_dir, "*.json"))

    for json_file in json_files:
        with open(json_file, "r", encoding="utf-8") as f:
            data = json.load(f)

        # 원본 이미지 파일명 가져오기
        image_name = data["description"]["image"]
        image_path = os.path.join(images_dir, image_name)

        if not os.path.exists(image_path):
            print(f"이미지 없음: {image_name}")
            continue  # 이미지가 없으면 스킵

        # 이미지 크기 저장
        img_width = data["description"]["width"]
        img_height = data["description"]["height"]

        # YOLO 포맷 라벨 파일명 설정
        yolo_label_path = os.path.join(output_dir, re.sub(r"\.(jpg|jpeg)$", ".txt", image_name, flags=re.IGNORECASE))

        # 바운딩 박스 변환 및 저장
        with open(yolo_label_path, "w") as yolo_file:
            for annotation in data["annotations"]["points"]:
                # 클래스 ID 가져오기, 없으면 -1 설정.
                class_id = CLASS_MAPPING.get(data["annotations"]["disease"], -1)

                if class_id == -1:
                    print(f"알 수 없는 클래스: {data['annotations']['disease']} (파일: {json_file})")
                    continue  # 매핑되지 않은 클래스는 스킵

                # 바운딩 박스 좌표 정규화
                xtl, ytl = annotation["xtl"], annotation["ytl"]
                xbr, ybr = annotation["xbr"], annotation["ybr"]

                x_center = (xtl + xbr) / 2.0 / img_width
                y_center = (ytl + ybr) / 2.0 / img_height
                width = (xbr - xtl) / img_width
                height = (ybr - ytl) / img_height

                # YOLO 포맷으로 저장
                yolo_file.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

        print(f"변환 완료: {yolo_label_path}")
```

### inference 후처리

#### detect된 object 없는 경우, (-1, 0, 0, 0) 텍스트 파일 추가

```python
# inference 결과 이미지와 라벨 디렉토리
inference_image_dir = '/content/output/inference_result'
inference_label_dir = '/content/output/inference_result/labels'

# 이미지 파일 이름 리스트
image_files = [f for f in os.listdir(inference_image_dir) if f.endswith(('.jpg', '.jpeg', '.JPG', '.JPEG'))]
# 라벨 파일 이름 리스트
label_files = [f for f in os.listdir(inference_label_dir) if f.endswith(('.txt'))]


for image_file in image_files:
    # 이미지 파일에 존재하는 이름의 라벨 파일 리스트
    label_file = image_file.replace('.jpg', '.txt').replace('.jpeg', '.txt').replace('.JPG', '.txt').replace('.JPEG', '.txt')

    # label_file에 있는 파일이 label_files에 없다면 (-1, 0, 0, 0) 텍스트 파일 생성
    if label_file not in label_files:
        label_path = os.path.join(inference_label_dir, label_file)

        # 클래스 ID -1로 라벨 파일 생성
        with open(label_path, 'w') as f:
            f.write("-1 0 0 0 0\n")

        print(f"(-1, 0, 0, 0) 라벨 파일 생성: {label_path}")
```
